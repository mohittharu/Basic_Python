```python
#Wap to print the circle as class and find out the perimeter and area
class Circle:
    radius = 5
    pi = 3.14
    area= pi*(radius*radius)
    perimeter = 2*pi*radius 
    print("Area of the circle is:",area)
    print("Perimeter of the circle is:",perimeter)
     
A1= Circle() 
print(A1)
```

```python
#Encapsulation
class A:
    x=5   #public
    
    _a= 10 # protected
    __b= 20    #private

    def Show(self):
        print("a= ",self._a)
        print("b= ",self.__b)
        print("x= ",self.x)


obj =A()
obj.Show()
#yeha public lai outside the class pani print garna milxa 
print("Outside the class ",obj.x)
print("out",obj._a)
print(obj.__b) # outside ma private print hudaina 
```

```python
class House:
    def __init__(self):
        print('Allowed.')

    def call(self):
        print('Ram are you there !!')

class Room1(House):
    def __init__(self):
        super().__init__()
        print('Any body in the house ')

    def greet(self):
        print('Hello')

class Room2(Room1):
    def __init__(self):
        print('I am here in Room2 !!')

    def Ask(self):
        print('Yes I am here!!')
    

room=Room1()
room.call()
```
