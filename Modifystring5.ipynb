# Modify string 

```python
a = "Hello , World !"

print(a.upper())      #upper case 

print(a.lower())       #lower case 

print(a.strip())     #strip() removes any whitespace 
```

```python
# String concatenation 

x = "Hello"
y = "World!" 

z = x + y 
print(z)
print(x + y)
```

```python
# f-string
age = 24
txt = f"My name is Mohit,and Age is {age}" 
print(txt)


# placeholder and Modifiers
price = 78
text = f"The price of pen is {price} dollars" 
print(text)
```
