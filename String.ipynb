# Accessing & indexing
- finding or access elements of iterable object using labeling/index

```python
word = "Dursikshya Education Network"
word
```

```python
print(word[2])
print(word[-1])     #negative indexing
print(word[10])
print(word[-10])    
```

# Slicing
- [startIndex : stopIndex)

```python
print(word[0:10])
print(word[11:20])
print(word[:20])
print(word[11:])
```

```python
print(word[0:20:3])
print(word[::-1])    #reverse string
print(word[::-2])
```

```python
'nepal'[0:3]
'nepal'[-5:-2]
```
