```python
# write a function to find the first positive  missing number in the given  listÂ Â ofÂ numbers

def first_missing_positive(nums):
    nums = set(nums)  # Remove duplicates for faster lookup
    smallest = 1

    while smallest in nums:
        smallest += 1

    return smallest

# Example usage:
numbers = [3, 4, -1, 1]
print("First missing positive number:", first_missing_positive(numbers))
    
```

```python
#write a function to find the smallest number in each list from a group of lists 
```

```python
def find_smallest(lst):
    rs = []
    for i in range (len(lst)):
        if(len(lst[i])) == 0:
            rs.append('Empty lst')

        else:
            smallest = lst[i][0]

            for j in range(len(lst[i])):
                if  lst[i][j]< smallest:

                    smallest = lst[i][j]
            rs.append(smallest)

        return rs

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(find_smallest(lst))    
            
```

```python
# write a function to repeat all vowels twice in a given string 
def vow_twice(s):
    v='aeiouAEIOU'
    result=''
    for i in s:
        if i in v:
            result += i *2
        else:
            result += i
    return result

vow_twice("Hello World")
```

```python
def repeat_vowels_twice(s):
    vowels = "aeiouAEIOU"
    result = ""
    for char in s:
        if char in vowels:
            result += char * 2  # repeat vowel twice
        else:
            result += char
    return result

# Example usage:
print(repeat_vowels_twice("Hello World"))  # Output: Heelloo Woorld
```

```python
# write a function to find the standard deviation using stdev() and round off upto 2 decimal 

import statistics
def find_standard_deviation(numbers):
    return round(statistics.stdev(numbers),2)

find_standard_deviation([1,2,3,4,5])
```

```python
#write a function to check if  a given  hand is  poker card is a full house 

from collections import Counter
def is_full_house(hand):
    counts = Counter(hand).values()

    return sorted (counts) == [2,3]
```
