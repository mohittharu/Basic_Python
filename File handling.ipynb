```python
# to open file 
-use open (filePath, mode )
-where filepath defines the file name with path where it is locate 
-mode defines how would like to open any files
-modes:
    -r: read only, it only read content of files. it generate error if file not found.
    -w:write only, it is only use for writing content to file .if filePath of filename doesnot 
        exist, it automatically creates it. it truncate existing content of file .

    -a:append only ,it only use for adding more content on existing file content.
    - '+' : if it apply on every mode, it gives read and write facility . 


# to close opened file.
-once you opne any file , must close it.
- variableName.close()

```

```python
# using with statement

-syntax :
    -with open(filePath, mode) as fileName:
        -content/method
-it is modern file handling technique.
- it helps to open file in different mode and close automatically .
```

```python
with open('my text.txt','w+')as fileWriter:
    text = input('Write about yourself:')
    fileWriter.write(text)
    fileWriter.seek(0)         #read write sangai garda use garni 
    content = fileWriter.read()
    print('About me !!')
    print('---------------')
    print(content)
```

```python
CSV file Handling
-comma seperated value

importing module named csv 
```

```python
import csv
```

```python
student_grades =[
    ['name','Assignment_1','assignment_2'],
    ['Gokul', 90,85],
    ['Mohit', 99,100],
    ['Anit', 80,99]
]
```

```python
#create new csv file
with open('grades.csv','w',newline='')as f:        #file banauxa 
    fileWriter= csv.writer(f)
    fileWriter.writerows(student_grades)
```

```python
#read csv file 
with open('grades.csv', 'r') as f:
    csv_reader = csv.reader(f)

    #reader is a iterator. The first row is often the header
    #next() it store header part and skip for other

    header = next(csv_reader)
    print(f'Header : {header}')


    #we can now loop over the remaining rows
    print('Student Data')
    for row in csv_reader:
        print(f'Name: {row[0]}, Assignment_1: {row[1]}, Assignment_2: {row[2]}')
```
