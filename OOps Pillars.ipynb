```python
##Access modifier 
1. Public :
- by default all member of class are public
- any member declared as pulbic can be accessed from outside of the class throught an objects
- we can also modify their values from outside of the class


2. Private:
- a double underscope '__'  makes member private as well as secure and member of the class which is declared as private  arenot accessible 
outside the class 
- but inside class it totallyu acessible 
- to access from outside the class we must create and method to access 
- for this we use getter and setter

3. Protected:
 - a singel underscope _ make member protected 
 - it assume that it cannot accessible from outside the class
 - it mainly used in inheritance mechanism 
 
```

```python
class AccessModifier:
    def __init__(self, a, b, c):
        self.a= a         # public
        self.__b= b          #private
        self._c= c          # protected

    def display(self):
        print('a:',self.a)
        print('b:',self.__b)
        print('c:',self._c)

    def __show(self):
        print('show all')
#method ma pani use hunxa access modifier not only in attribute 
    def _greet(self):
        print('Hello ')       
        
```

```python
access = AccessModifier(1, 2, 3)
access.display()
print(access.a)

access.a=10
print('access is :',access.a)
```

```python
class BankAccount:
    def __init__(self, accNo, balance):
        self.accNo= accNo
        self.__balance= balance   #private attribute

    def __deposit(self,amt):
        if amt>0:
            self.__balance += amt
            return f'Deposited {amt} , new Balance: {self.__balance}'
        return 'Invalid Amount'

    def getBalance(self):
        return self.__balance

    def operate(self,amt):
        return self.__deposit(amt)
        
```

```python
acc1= BankAccount(101,5000)
acc1
```

```python
acc1.getBalance()
acc1.operate(15000)
```

```python
SINGLE INHERITANCE 
```

```python
class Person:
    def __init__(self, name, age, address, phone):
        self.name= name 
        self.age = age
        self.address = address
        self.phone = phone


    def display(self):
        print('***** General Details *******')
        print('Name: ',self.name)
        print('Age:',self.age)
        print('Address:',self.address)
        print('Phone:',self.phone)

    def show(self):
        print('Hello miss')

    #derived class
class Student(Person):
    def __init__(self, name, age, address, phone, faculty, roll_no):
       # super() method used for access parent's member and use only for single inheritance 
        super().__init__(name, age, address, phone)   # mathi  pathauna super use gareko
        self.faculty = faculty
        self.roll_no = roll_no

    def display(self):
        super().display()
        print('*** Academic Detail******')
        print('Faculty:',self.faculty)
        print('Roll_no:',self.roll_no)
    
```

```python
std1 = Student('Ram', 19, 'KTM', '9800574974', 'Science','107')
std1.display()
std1.show()     #method call garey ko 
std1.name       #direct call garna pauxa 
```

```python
class Instructor:
    def __init__(self):
        
        self.id = ''
        self.name = ''
        self.post = ''

    def getData(self):
        print('***Enter Instructor Details *****')
        print('_'*10)
        self.id = input('Enter I-ID')
        self.name = input('Enter I-Name')
        self.post = input('Enter I-Post')

    def show(self):
        print('_'*10)
        self.id =(' I-ID',self.id)
        self.name = (' I-Name',self.name)
        self.post = (' I-Post',self.post)

class Student:
    def __init__(self):
        
        self.sid = ''
        self.sname = ''
        self.level = ''

    def getData(self):
        print('***Enter Student Details *****')
        print('_'*10)
        self.sid = input('Enter S-ID')
        self.sname = input('Enter S-Name')
        self.level = input('Enter S-Level')

    def show(self):
        print('_'*10)
        self.sid =(' S-ID',self.sid)
        self.sname = (' S-Name',self.sname)
        self.level = (' S-Level',self.level)

#multiple inheritance 
class TA(Instructor, Student):
    def __init__(self):
        self.prof = ''

    def getData(self):
        Instructor.getData(self)
        Student.getData(self)
        print("*******Enter Professor's Detail****")
        print('_'*10)
        self.prof = input("Enter Prof Name:")

    def show(self):
       Instructor.show(self)
       Student.show(self)
       print("Professor Name:",self.prof)
       
    
```

```python
# Pplymorphism 

-poly:many,morphism:shape
-having same name but different functionality 
```

```python
class Bird:
    def fly(self):
        print('Bird can fly')

    def swim(self):
        print('Bird cannot swim')

class Parrot(Bird):
    def fly(self):
        super().fly()
        print('Parrot can fly')

class Penguin(Bird):
    def fly(self):
        print('Penguin canot fly')

    def swim(self):
        super().fly()
        print('But Penguin can swim')
        
```

```python
parrot = Parrot()
parrot.fly()
parrot.swim()
```

```python
Penguin = Penguin()
Penguin.fly()
Penguin.swim()
```
