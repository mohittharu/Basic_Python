```python
#list of purchase amount given them discount on the basis of purchase amount 1000 no dis, upto 5k 2% , upto 10k 5%,more than 10k 10%

```

```python
def discount(amt):
    if amt>=10000:
        return amt-amt*0.1
    elif amt>=5000:
        return amt-amt*0.05
    elif amt>=1000:
        return amt-amt*0.02
    else:
        return amt
        
purchase_amount = [ 800, 1000, 300, 1500, 5010, 15000, 4000, 7500]
after_discount= list(map(discount, purchase_amount))
after_discount
```

```python
#factorial 
def factorial(n): # defining function #5          4           3           
    if n<=1:
        return 1
    return n* factorial(n-1)   #5 *4 * 3 * 2 *1 *1 *1

factorial(5)  #calling function 
```

```python
# fibonacci series 
def fibo(n):

    # define function 
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)

num=int(input('Enter number range:'))
for i in range(num):
    print(fibo(i),end=' ')



```

```python
fibo(3)    #calling function 
```
