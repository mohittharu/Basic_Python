```python
# Blocks for Error Handling 
-try block , ler you test a block of code for possible errors.
-except block, lets you handle the errors.
-else block, lets you excecute the code when there is no errors.
-finally block , lets you execute block of code regardless of the result from try and except block .
```

```python
def fileEditor(path='', content='', mode=''):
    try:
        fileObj = open(path, mode)
        try:
            fileObj.write(content)
        except:
            print('File format is not writtenable.you must pass mode to(w,a).')
            mode = input('Enter correct required mode(w,a) to write content.')
            fileEditor(path, content, mode)
            
            
    except fileNotFoundError:
        print('FileName not found')
        path = input('Enter File Name/path')
        fileEditor(path, content, mode)

    else:
        print('Added Content successfully')
    finally:
        fileObj.close()
        print('File Closed')
```

```python
path = input('Enter File Name/Path')
content = input('Enter Content you want to write.')
mode = input('Select required mode for editor(r,w,a):')
fileEditor(path, content, mode)
```
