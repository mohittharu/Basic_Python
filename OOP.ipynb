```python
#wap by creating class circle and find the area and the perimeter 
class Circle:
    pi=3.14
    def __init__(self,radius,pi):
        self.r= radius
        self.pi=pi

    def area(self):
        self.Area= self.pi * (self.r*self.r)
        print("Area of the circle is:",self.Area)

    def perimeter(self):
        self.Perimeter= 2 * self.pi * self.r
        print("Perimeter of the circle is:",self.Perimeter)

a=Circle(5,3.14)
a.area()
a.perimeter()

        
    
```

```python
# define class
class Account:
    minBalc = 1000              # class variable(yo hamro global ho kina ki yo scope bahira define gariyeko xa )
    accType = 'Saving'

    def __init__(self, accNo, name, balance):       # constructor
        self.accNo = accNo           # instance variable (hya assign garne ho )
        self.name = name
        self.balance = balance
 
    def details(self):                    # instance method   (sabai ma use nahuna ni sakxxa )
        print('Account Details')
        print('AccType:', Account.accType)
        print('AccNo:',self.accNo)
        print('Name:', self.name)
        print('Balance:', self.balance)

    @classmethod
    def showAccount(cls):                      # class method
        print('Account Type:', cls.accType)
        print('Minimum Balance for this type:', Account.minBalc)

    def deposit(self,amt):
        self.balance+=amt
        print(f'Your account {self.accNo} has been credited by Rs.{amt}. Your final balance is {self.balance}')

    def withdraw(self, amt):
        res = self.checkBalance(amt, self.balance, self.minBalc)
        if res==-1:
            print('Insufficient Balance.')
        elif res==0:
            print('Amount should be greater than 1000')
        else:
            self.balance -= amt
            print(f'Your account {self.accNo} has been debited by Rs.{amt}. Your final balance is {self.balance}')

    @staticmethod     #farak pardaina direct garna pauni ho 
    def checkBalance(amt, balc, minB):
        if amt>balc:
            return -1
        elif amt<=minB:
            return 0
        else:
            return 1
    
# object creating and operation
acc1 = Account('001', 'Gokul', 10000)
acc1.details()
# acc1.showAccount()
# Account.showAccount()

acc1.deposit(5000)
acc1.withdraw(1001)
```
